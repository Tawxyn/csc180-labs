```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("carprices4.csv")
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
# Drop redudant Car Model column, and dummy variable trap
final = pd.concat([df, dummies], axis=1)
final = final.drop(["Car Model", dummies.columns[-1]], axis=1)
final
```

```python
X = final.drop(["Sell Price($)"], axis=1)
X
```

```python
y = final["Sell Price($)"]
y
```

```python
# Linear regression score
model = LinearRegression().fit(X, y)
model.score(X,y)
```

```python
# Prediction for 4 year old mercedez benz with 45,000 miles. BMW > Mercedez
X_new = pd.DataFrame(
    [[45000, 4, 0, 0]],
    columns=['Mileage', 'Age(yrs)', 'Audi A5', 'BMW X5']
)
pred = model.predict(X_new)
print("Prediction:", pred)
```
